### Renaming the Column name "movie_id" to "id" in credits dataset

```python
credits=credits.rename(columns={"movie_id":"id"})
```

```python
credits.head()
```

### Merging the dataset "credits" into the "movies_df" based on the key value of "id" column

```python
movies_df_merged=movies_df.merge(credits,on='id')
```

```python
movies_df_merged.shape
```

```python
movies_df_merged.head()
```

# Dropping the unnecessary columns from the Dataset movies_df_merged

```python
movie_df_cleaned=movies_df_merged.drop(columns=['homepage','title_x','title_y','status','production_countries'])
```

```python
movie_df_cleaned.shape
```

```python
movie_df_cleaned.head()
```

```python
movie_df_cleaned.info()
```

```python
sns.heatmap(movie_df_cleaned.isnull())
```

# Weighted Average formula : W=(Rv+Cm)/(v+m)
where : 
    W= Weighted Rating
    R=average of all the ratings for a movie as a number from 0 to 10(mean)=Rating
    v=number of votes for the movie(votes)
    m=minimum number of votes required to be listed in the Top 250 movies is 250 votes
    C=the mean

```python
C=movie_df_cleaned['vote_average'].mean()
```

```python
C
```

```python
m=movie_df_cleaned['vote_count'].quantile(0.70)
```

```python
m
```

```python
R=movie_df_cleaned['vote_average']
```

```python
v=movie_df_cleaned['vote_count']
```

```python
weighted_average=(R*v+C*m)/(v+m)
```

```python
weighted_average=pd.DataFrame(weighted_average,columns=['weighted_average'])
```

```python
weighted_average
```

```python
movie_df_cleaned['weighted_average']=weighted_average
```

```python
movie_df_cleaned.head()
```

```python
movies_sorted_by_wa=movie_df_cleaned.sort_values('weighted_average', ascending=False)
```

```python
movies_sorted_by_wa[['original_title','vote_average','vote_count','weighted_average','popularity']].head(20)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=movies_sorted_by_wa['weighted_average'].head(10),y=movies_sorted_by_wa['original_title'].head(10),palette='cool')
plt.xlabel('Weighted Average Score',weight='bold')
plt.ylabel('Movie Titles',weight='bold')
plt.title('Top Rated Movies based on Ratings given by Viewers',weight='bold')
plt.savefig('Top10movies_by_weightedAverage.png')
```

### Now let's sort and plot the movies based on the Popularity values

```python
movie_popularity=movie_df_cleaned.sort_values('popularity',ascending=False)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=movie_popularity['popularity'].head(10),y=movie_popularity['original_title'].head(10),palette='cool')
plt.xlabel('Weighted Average Score',weight='bold')
plt.ylabel('Movie Titles',weight='bold')
plt.title('Top Rated Movies based on Ratings given by Viewers',weight='bold')
plt.savefig('top10movies_by_popularity.png')
```

## Now we will  import the MinMaxScaler function from SKlearn library to scale the values of 'wighted_average' and 'popularity' columns on the same scale
And here is the formula MinMaxScaler() use internally to convert the values into the range of 0 to 1 :



```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaling=MinMaxScaler()
```

```python
scaled_weight_popularity=scaling.fit_transform(movie_df_cleaned[['weighted_average','popularity']])
```

```python
scaled_weight_popularity.shape
```

```python
scaled_weight_popularity=pd.DataFrame(scaled_weight_popularity,columns=['weighted_average','popularity'])
```

```python
scaled_weight_popularity.head()
```

```python
movie_df_cleaned[['normalized_weight','normalized_popularity']]=scaled_weight_popularity
```

```python
movie_df_cleaned
```

```python
movie_df_cleaned['score']=movie_df_cleaned['normalized_weight']*0.5+movie_df_cleaned['normalized_popularity']*0.5
```

```python
movies_sortedBy_score=movie_df_cleaned.sort_values('score',ascending=False)
```

```python
movies_sortedBy_score
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=movies_sortedBy_score['score'].head(10),y=movies_sortedBy_score['original_title'].head(10))
plt.xlabel('Score of Movies',weight='bold')
plt.ylabel('Titles of Movies',weight='bold')
plt.title('Top rated movies based on Popularity and Weighted_Average',weight='bold')
```
